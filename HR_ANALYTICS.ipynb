## Plot a correlation map for all numeric variable Overtime

```python
# Encode OverTime as numeric
df['OverTime'] = df['OverTime'].map({'Yes': 1, 'No': 0})

# Select numeric variables again
numeric_variables = df.select_dtypes(include='number')

# Calculate correlation matrix
correlation_matrix = numeric_variables.corr()

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap for Numeric Variables (including OverTime)')
plt.show()

```

```python
correlation_matrix = df.select_dtypes(include='number').corr()

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix (Numeric Only)')
plt.show()

```
